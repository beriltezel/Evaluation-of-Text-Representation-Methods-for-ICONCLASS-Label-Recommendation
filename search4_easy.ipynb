```python
from iconclass import init
```

```python
ic = init()  # root node (Notation)
```

```python
def walk(node):
    """Yield node and all descendants."""
    yield node
    for child in node:  # node is iterable → children
        yield from walk(child)
```

```python
def search_iconclass(query, lang="en"):
    query = query.lower()
    results = []

    for node in walk(ic):
        # get notation code
        notation = repr(node).split()[0] 

        # check in notation
        if query in notation.lower():
            results.append(node)
            continue

        # check in label
        try:
            label = node(lang)
        except:
            label = ""
        if query in label.lower():
            results.append(node)

    return results
```

```python
results = search_iconclass("lion")

for r in results:
    print(repr(r), "→", r())
```
